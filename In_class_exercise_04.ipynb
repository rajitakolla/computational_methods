```python
import urllib.request
legal_case_df = urllib.request.urlopen("https://raw.githubusercontent.com/unt-iialab/INFO5731_FALL2020/master/In_class_exercise/01-05-1%20%20Adams%20v%20Tanner.txt")


#iterate over the webpage to extract the data
required_values=list()
for data in legal_case_df:
  values = data.decode("utf-8").replace("\r\n",'')
  if values:
    required_values.append(values)

#question 1
print("The total number of sentences : ",len(required_values))
# print("\n---------------------------------------------------------------------------------------------\n")
# for i in required_values:
#   print(i)
```

```python
required_df = pd.DataFrame(required_values,columns=['data'])
import re,pandas as pd
pd.set_option("display.max_rows", None, "display.max_columns", None)

#number of words in each sentence
res1 = required_df['data'].apply(lambda x: len(x.split(" "))).to_frame()
word_count_df = pd.concat([required_df,res1.rename(columns ={'data':'word_count'})],axis=1)
print(word_count_df)
```

```python
#count the number of characters in senetnces
#Here I am counting specifically characters excluding punctuation
res2_1 = required_df['data'].apply(lambda x:''.join([i for i in x if i not in string.punctuation])).to_frame().rename(columns = {'data':'filtered_data'})
res2 = res2_1['filtered_data'].apply(lambda x: len(x.replace(" ",""))).to_frame().rename(columns = {"filtered_data":"character count"})
char_count_df = pd.concat([word_count_df,res2],axis=1)
char_count_df
```

```python
#Average word length
res3 = required_df['data'].apply(lambda x: len(x.replace(" ",""))/len(x.split()) if len(x.split())!=0 else 0).to_frame().rename(columns={"data":"avg_wrd_cnt"})
average_word_count = pd.concat([char_count_df, res3],axis=1)
average_word_count
```

```python
#count of numeric characters
res4 = required_df['data'].apply(lambda x: sum(c.isdigit() for c in x)).to_frame().rename(columns={"data":"numeric_columns"})
numeric_values_count_df = pd.concat([average_word_count,res4],axis=1)
numeric_values_count_df
```

```python
#count the number of special characters
def count_of_special_chars(val):
  count = 0
  for i in val:
    if i.isdigit() or i.isalpha() or (i == ' '):
      count=count
    else:
      count=count+1
  return count
  pass
res5 = required_df['data'].apply(lambda x: count_of_special_chars(x)).to_frame().rename(columns={"data":"spl_chars"})
special_char_df = pd.concat([numeric_values_count_df,res5],axis=1)
special_char_df
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
#count the number of stop words
list_of_words = set(stopwords.words('english')) 
res6 = required_df['data'].apply(lambda x: len([i for i in x.split() if i in list_of_words])).to_frame().rename(columns={"data":"count_of_stop_words"})
stop_words_df = pd.concat([special_char_df,res6],axis=1)
stop_words_df
```

```python
#top 15 frequent words
frequent_words = list(pd.Series(' '.join(stop_words_df['data']).split()).value_counts()[:15].index)
#top 15 rare words
rare_words = list(pd.Series(' '.join(stop_words_df['data']).split()).value_counts()[-15:].index)
```

```python
#count the number of upper case words
res7 = required_df['data'].apply(lambda x: sum(i.isupper() for i in x)).to_frame().rename(columns={"data":"upper_case_count"})
uppper_case_count = pd.concat([stop_words_df,res7],axis=1)
# uppper_case_count
```

```python
res8 = required_df['data'].apply(lambda x:''.join([i for i in x if i not in string.punctuation])).to_frame().rename(columns={"data":"punctuations"})

punctuation_removal = pd.concat([uppper_case_count,res8],axis=1)
punctuation_removal
```

```python
#lowercasing
res9 = punctuation_removal['punctuations'].apply(lambda x: x.lower()).to_frame().rename(columns={"punctuations":"lower_case"})
lower_casing = pd.concat([punctuation_removal,res9],axis=1)
lower_casing
```

```python
#stopwords removal
res10 =  lower_casing['lower_case'].apply(lambda x:' '.join([i for i in x.split() if i not in list_of_words])).to_frame().rename(columns={"lower_case":"stop_words"})
stop_words = pd.concat([lower_casing,res10],axis=1)
stop_words
```

```python
#frequent words removal 
res11 = stop_words['stop_words'].apply(lambda x: ' '.join([i for i in x.split() if i not in frequent_words])).to_frame().rename(columns={"stop_words":"frequent_words"})
frequent_words = pd.concat([stop_words,res11],axis=1)
frequent_words
```

```python
#rare words removal
res12 = frequent_words['frequent_words'].apply(lambda x: ' '.join([i for i in x.split() if i not in rare_words])).to_frame().rename(columns={"frequent_words":"rare_words"})

rare_words_removal =pd.concat([frequent_words,res12],axis=1)
rare_words_removal
```

```python
#spelling correction
from textblob import TextBlob
res13 = rare_words_removal['rare_words'].apply(lambda x: str(TextBlob(x).correct())).to_frame().rename(columns={"rare_words":"spell_check"})
spell_correction = pd.concat([rare_words_removal,res13],axis=1)
spell_correction
```

```python
nltk.download('punkt')
```

```python
#tokenisation
from nltk import word_tokenize
res14 = spell_correction['spell_check'].apply(lambda x: word_tokenize(str(x))).to_frame().rename(columns={"spell_check":"tokenization"})
tokenisation = pd.concat([spell_correction,res14],axis=1)
tokenisation
```

```python
#stemming
from nltk.stem import PorterStemmer 
from nltk.tokenize import word_tokenize 
   
ps = PorterStemmer() 
```

```python
# stemming
res15 = tokenisation['tokenization'].apply(lambda x: ' '.join([ps.stem(i) for i in x])).to_frame().rename(columns={"tokenization":"stemming"})
stemming = pd.concat([tokenisation,res15],axis=1)
stemming
```

```python
import textblob
res16 = stemming['tokenization'].apply(lambda x: ' '.join([textblob.Word(i).lemmatize() for i in x])).to_frame().rename(columns={"tokenization":"lemmitization"})
lemmatization = pd.concat([stemming,res16],axis=1)
lemmatization
```
