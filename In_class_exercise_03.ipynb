```python
# You code here (Please add comments in the code):

import urllib.request
import requests 
from bs4 import BeautifulSoup
import pandas as pd
#extarct the webpage requests
page_info = urllib.request.urlopen("https://www.timeanddate.com/astronomy/canada")
#create a Beautifulsoup class to parse the page
soup = BeautifulSoup(page_info)
#print(soup.prettify)
#extract table data
table_data = soup.findAll('table',attrs={'class':'zebra fw tb-sm zebra'})
```

```python
#exract various areas in canada
areas_list = list()
for value in table_data:
  #print(f.find_all('a'))
  for area_name in value:
    if(area_name.find('a')!=None):
      areas_list.append(area_name.find('a').get_text())

print(areas_list)
```

```python
from datetime import datetime
#fecth system day, month and year
currentDay = datetime.now().day
currentMonth = datetime.now().month
currentYear = datetime.now().year
```

```python
#get the sunset and sunrise time for each location 
print("date             sunrise         sunset          daylength    day-diff  current-time            milliseconds")
days = [i for i in range(currentDay,currentDay+7)]
# print(days)
#iterate over areas
for i in areas_list:
  print("State name: ",i)
  try:
    #dynamic access of webpage
    page = urllib.request.urlopen("https://www.timeanddate.com/astronomy/canada/"+i.replace(" ",""))
    soup = BeautifulSoup(page)

    #print(soup.prettify)
    values = soup.findAll("table", {"class": "zebra tb-sm tb-hover fw va-m"})
    #print(values)
    #iterate over the values
    for value in values:
      #iterate over days
      for day in days:
        day_data = value.findAll("tr",{"data-day":day})
        #print(day_data)
      
        for i in day_data:
          class_names = ["c sep","sep c","c tr sep-l","c tr sep","c sep-l","tr sep"]
          r = list()
          r.append(str(currentMonth)+"/"+str(day)+"/"+str(currentYear))
          for class_name in class_names:
            if(i.find("td",{"class":class_name})!=None):
              r.append(i.find("td",{"class":class_name}).get_text())
            else:
              r.append(None)
          #display the result
          [print(a+"\t", end =" ") for a in r]
          print("")
  except:
    print("Invalid URL")
        
```
